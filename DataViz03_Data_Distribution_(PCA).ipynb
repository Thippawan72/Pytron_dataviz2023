### เริ่มทำ PCA

sklearn -> scikit-learn เป็น package ที่รวบรวม function การทำ Data Science - Machine Learning - Data Mining เอาไว้ให้เราใช้แบบไม่ต้องเขียนเอง

import

```python
from sklearn.decomposition import PCA
```

define

```python
pca = PCA()
```

fit-transform ตัวเเปรทั้ง 4 ตัว

```python
thisdata = example_df.iloc[:,:-1]
thisdata
```

```python
new_axis = pca.fit_transform(thisdata)
```

```python
new_axis.shape
```

```python
new_axis
```

```python
PCAdf = pd.DataFrame.from_records(new_axis, columns = ['PCA1','PCA2','PCA3','PCA4'])
PCAdf
```

```python
pca.explained_variance_ratio_ #ค่าความเเปรปรวนมากยิ่งอธิบายข้อมูลได้ดี เช่น ตัวเเปรที่ 1 สามารถอธิบายข้อมูลได้ถึง 92 เปอร์เซ็นต์ จึงสามารถเลือกใช้เเกนเดียวก็ได้
```
