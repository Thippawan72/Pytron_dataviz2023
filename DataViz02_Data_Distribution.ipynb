1. เขียน function ที่รับ input เป็น output ของ boxplot เเล้ว แสดงค่า min,max,q1,q2,q3

```python
x = plt.boxplot(df['SepalWidth'])
```

```python
def box_vals(x):
    q1 = x['boxes'][0].get_ydata()[0]
    q2 = x['medians'][0].get_ydata()[0]
    q3 = x['boxes'][0].get_ydata()[2]
    iqr = q3-q1
    vmin = min(x['whiskers'][0].get_ydata())
    vmax = max(x['whiskers'][1].get_ydata())
    return  print(f'Q1 เท่ากับ {q1}\nQ2 เท่ากับ {q2}\nQ3 เท่ากับ {q3}\nIQR เท่ากับ {iqr}\nMin เท่ากับ {vmin}\nMax เท่ากับ {vmax}')
```

```python
box_vals(x)
```

```python
def detail_boxreturn_v2(boxreturn, vert=True): # สร้าง list มารับ ##vert=True เเนวตั้ง
    if vert ==True:
        q1 = boxreturn['boxes'][0].get_ydata()[0]
        q2 = boxreturn['medians'][0].get_ydata()[0]
        q3 = boxreturn['boxes'][0].get_ydata()[2]
        minv = min(boxreturn['whiskers'][0].get_ydata())
        maxv = max(boxreturn['whiskers'][1].get_ydata())
    else: # vert == False
        q1 = boxreturn['boxes'][0].get_xdata()[0]
        q2 = boxreturn['medians'][0].get_xdata()[0]
        q3 = boxreturn['boxes'][0].get_xdata()[2]
        minv = min(boxreturn['whiskers'][0].get_xdata())
        maxv = max(boxreturn['whiskers'][1].get_xdata())

    print(f'min = {minv}, max = {maxv}, q1 = {q1}, q2 = {q2}, q3 = {q3}')
```
