```python
import pandas as pd
import os
os.chdir('/content/drive/Shareddrives/Exam')
```

```python

Tran_2019_2 = '/content/drive/Shareddrives/Exam/POS_transactions_20190213-20190226.csv'
Op_2019_2 =  '/content/drive/Shareddrives/Exam/POS_operator_logs_20190213-20190226.csv'

Tran_2019_3 = '/content/drive/Shareddrives/Exam/POS_transactions_20190328-20190410.csv'
Op_2019_3 ='/content/drive/Shareddrives/Exam/POS_operator_logs_20190328-20190410.csv'

# Read the CSV file into a DataFrame

df_T2019_2 = pd.read_csv(Tran_2019_2)
df_O2019_2 = pd.read_csv(Op_2019_2)
df_T2019_3 = pd.read_csv(Tran_2019_3)
df_O2019_3 = pd.read_csv(Op_2019_3)
```

#Month 2

```python
import pandas as pd

# Assuming 'df_T2019_3' is your DataFrame
df_T2019_2['PaymentMethod'] = 'other'  # Initialize the new column with a default value

# Set 'PaymentMethod' to 'cash' when 'TNcash' is True and 'TNcard' is False
df_T2019_2.loc[(df_T2019_2['TNcash'] == True) & (df_T2019_2['TNcard'] == False), 'PaymentMethod'] = 'cash'

# Set 'PaymentMethod' to 'card' when 'TNcash' is False and 'TNcard' is True
df_T2019_2.loc[(df_T2019_2['TNcash'] == False) & (df_T2019_2['TNcard'] == True), 'PaymentMethod'] = 'card'
```

```python

df_T2019_2['WorkstationGroupID'].replace({1: 'Service', 8: 'Selfservice'}, inplace=True)
```

```python
import pandas as pd

# Assuming 'df_T2019_3' is your DataFrame
# Convert 'BeginDateTime' and 'EndDateTime' columns to datetime objects
df_T2019_2['BeginDateTime'] = pd.to_datetime(df_T2019_3['BeginDateTime'])
df_T2019_2['EndDateTime'] = pd.to_datetime(df_T2019_3['EndDateTime'])

# Calculate the time difference in seconds and create a new column 'TimeDifference'
df_T2019_2['TimeDifference'] = (df_T2019_2['EndDateTime'] - df_T2019_2['BeginDateTime']).dt.total_seconds()
```

- Break time

```python

grouped_df11 = df_T2019_2.groupby('OperatorID')['BreakTime'].sum().reset_index()

# Rename the 'BreakTime' column to 'TotalBreakTime' in the grouped DataFrame
grouped_df1 = grouped_df11.rename(columns={'BreakTime': 'TotalBreakTime'})
grouped_df1
```

- Tran time

```python

grouped_df111 = df_T2019_2.groupby('OperatorID')['TranTime'].sum().reset_index()

# Rename the 'BreakTime' column to 'TotalBreakTime' in the grouped DataFrame
grouped_df11 = grouped_df111.rename(columns={'TranTime': 'TotalTranTime'})
```

- group Break and Tran

```python
merged_grouped_df111 = pd.merge(grouped_df1, grouped_df11, on='OperatorID', how='inner')
```

```python
merged_grouped_df1111 = pd.merge(grouped_df111, df_O2019_2, on='OperatorID', how='inner')

```

```python

translations = {
    'OperatorSignOn': 'เริ่มทํางาน',
    'OperatorSignOff': 'จบการทํางาน',
    'OperatorLock': 'พักงานชั่วคราว',
    'OperatorUnLock': 'กลับมาทํางานต่อหลังจากพัก'
}

merged_grouped_df1111['TranslatedItems'] = merged_grouped_df1111['Items'].replace(translations)
merged_grouped_df1111
```

```python
merged_grouped_df1111.to_csv('Break_Tran_2.csv', index=False)
```
