```python
BKK = data_df[data_df['province_of_onset']=='กรุงเทพมหานคร'] # เลือกดู กรุงเทพ
BKK.head() # ต้องออกมา 5 เเถว
```

```python
KKC = data_df[data_df['province_of_onset']=='ขอนแก่น'] # เลือกดู ขอนเเก่น
KKC.head() # ต้องออกมา 5 เเถว
```

### ถ้าเอาตารางไม่เหมือนกันมา concat กัน

```python
pd.concat([BKK2,KKC2]) # ถ้าคอลัมน์เป็นชื่อเดียวกันจะ สลับที่ให้เลย
```

```python
pd.concat([BKK2,KKC3]) # มันก็จะพบค่าว่างในคอลัมน์ที่ไม่เหมือนกัน
```

```python
KKC4 = data_df[data_df['province_of_onset']=='ขอนแก่น'][['age','nationality']] # เลือกดู ขอนเเก่น เเต่เลือกเฉพาะคอลัมน์ 'age','nationality'
KKC4.head()
```

## ต่อเเกน X

```python
personal_data = data_df[['No.','sex','age','nationality']]
personal_data.head() # ตารางข้อมูลส่วนบุคคล
```

```python
covid_data = data_df[['No.','announce_date','province_of_isolation','province_of_onset','risk']].sample(frac=1) # .sample ไว้สลับเเถวเเบบสุ่ม
covid_data.head() # ข้อมูลโควิด
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data) # นำ no.1 มาเเปะ จะเอาตารางสองมาเเปะ คือ covid_data มาต่อ
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data) # นำ no.1 มาเเปะ จะเอาตารางสองมาเเปะ คือ personal_data มาต่อ
personalcovid_data.head()
```
